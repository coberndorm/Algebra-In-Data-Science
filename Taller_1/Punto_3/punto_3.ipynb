Cast points into the shape of the $l_p$ distance

```python
n = 2
p = 1
r = 1
l_p_distance = lambda x, y, p: np.sum(np.abs(x-y)**p)**(1/p)
points = np.random.random((5,n))
points = [r * point/l_p_distance(point, np.zeros(n), p) for point in points]

points
```

Create points inside a $l_p$ distance

```python
# metodo a
points = []
n = 5
p = 1
r = 1

while len(points) < 5:
    point = np.random.random((n))*r
    if l_p_distance(point, np.zeros(n), p) <= r:
        points.append(point)

points
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the L_p norm function
def lp_norm(x, y, p):
    return (np.abs(x)**p + np.abs(y)**p)**(1/p)

# Create a grid of points in 2D space
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Choose the value of p (e.g., p=1 for L1 norm, p=2 for L2 norm)
p = 100

# Calculate the L_p norm for each point in the grid
Z = lp_norm(X, Y, p)

# Create a contour plot of the L_p norm
plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'L{p} Norm Contour Plot')
plt.colorbar(label=f'L{p} Norm')
plt.grid(True)
plt.axis('equal')
plt.show()

```
