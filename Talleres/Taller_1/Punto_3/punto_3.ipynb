## Parte b.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_points(points_x, points_y):
    # Generar puntos aleatorios en el plano
    def calculate_minkowski_distance(x, y, p):
        return (abs(x)**p + abs(y)**p)**(1/p)

    # Par√°metros de la bola de Lp
    r = 1.0  # Radio de la bola
    p_values = [1, 2, 3, 4, 10]  # Valores de p para las diferentes distancias Lp

    # Crear una malla para trazar la frontera de la bola
    grid_size = 1000
    x = np.linspace(-r, r, grid_size)
    y = np.linspace(-r, r, grid_size)
    X, Y = np.meshgrid(x, y)

    # Graficar la bola de Lp y los puntos aleatorios
    plt.figure(figsize=(8, 6))

    for p in p_values:
        Z = calculate_minkowski_distance(X, Y, p)
        plt.contour(X, Y, Z, levels=[r], colors=[np.random.rand(3,)])

    plt.scatter(points_x, points_y, s=5, color='blue', alpha=0.5, label='Puntos aleatorios')
    plt.xlim(-r, r)
    plt.ylim(-r, r)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Bola de $L_p$ en el plano $\mathbb{R}^2$')
    plt.legend()
    plt.grid(True)
    plt.show()
```

Puntos en la frontera $L_p$

```python
points_amount = 100 # Define la cantidad de muestra de puntos

n = 2 # Define la dimensi√≥n del espacio (en este caso, 2D)
p = 1 # Define el orden (p) de la distancia (L_p)
r = 1 # Define el radio de la esfera

# Define una funci√≥n lambda que calcula la distancia (L_p) entre dos puntos (x) e (y)
l_p_distance = lambda x, y, p: np.sum(np.abs(x-y)**p)**(1/p)

# Genera una matriz de puntos aleatorios en el intervalo [-r, r] en ambas dimensiones
points = np.random.random((points_amount,n)) * 2 * r - r

# A continuaci√≥n, aplica una transformaci√≥n a los puntos generados
# para que est√©n en el borde de la esfera de radio (r) utilizando la distancia (L_p).
# Esto se hace dividiendo cada punto por su distancia (L_p) al origen y luego multiplic√°ndolo por (r).
points = [r * point/l_p_distance(point, np.zeros(n), p) for point in points]

plot_points([point[0] for point in points], [point[1] for point in points])
# points # Descomentar para ver los puntos generados
```

Puntos adentro de la frontera $L_p$

```python
points_size = 100 # Define el tama√±o de la muestra de puntos

points = [] # Inicializa una lista vac√≠a para almacenar los puntos generados
n = 2 # Define la dimensi√≥n del espacio (en este caso, 2D)
p = 1 # Define el orden (p) de la distancia (L_p)
r = 1 # Define el radio de la esfera

# Define una funci√≥n lambda que calcula la distancia (L_p) entre dos puntos (x) e (y)
l_p_distance = lambda x, y, p: np.sum(np.abs(x-y)**p)**(1/p)

# Utiliza un bucle 'while' para generar puntos hasta alcanzar el tama√±o deseado
while len(points) < points_size:
    # Genera un punto aleatorio en el intervalo [-r, r] en ambas dimensiones
    point = np.random.random((n)) * 2 * r - r
    
    # Comprueba si la distancia (L_p) de este punto al origen es menor o igual a (r)
    if l_p_distance(point, np.zeros(n), p) <= r:
        # Si cumple el criterio, agrega el punto a la lista de puntos generados
        points.append(point)

plot_points([point[0] for point in points], [point[1] for point in points])
# points # Descomentar si se desea ver la lista de puntos generados
```
