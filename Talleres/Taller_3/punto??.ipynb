```python
import numpy as np
import matplotlib.pyplot as plt
```

En un modelo SIR suponga que inicialmente 100% de la poblacion es susceptible. Proyecte 2.5 periodos de tiempo las propcrciones del SIR

Parametros:

$$
\begin{bmatrix}
0.9 & 0 & 0.2 \\
0.1 & 0.85 & 0 \\
0 & 0.15 & 0.8
\end{bmatrix}
\begin{bmatrix}
S\\
I\\
R
\end{bmatrix}
$$

```python
transition_matrix = np.array([[0.9, 0,    0.2],
                              [0.1, 0.85, 0  ],
                              [0,   0.15, 0.8]])

SIR = np.array([1000,0,0])
poblacion = [SIR]
for i in range(25):
    SIR = np.dot(transition_matrix,SIR)
    poblacion.append(SIR)
```

```python

plt.plot(np.array(poblacion))
plt.legend({"Susceptibles", "Infectados", "Recuperados"})
```
